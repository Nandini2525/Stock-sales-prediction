```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
dataset=pd.read_csv("/content/NFLX.csv")
```

```python
dataset=dataset[['Close']]
dataset.head()
```

```python
future_days=50
```

```python
dataset['Prediction']=dataset[['Close']].shift(-future_days)
dataset.tail(4)
```

```python
x=np.array(dataset[['Close']])
x=x[:dataset.shape[0]-future_days]
x
```

```python
y=np.array(dataset["Prediction"])[:-future_days]
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
x_train
```

```python
from sklearn.tree import DecisionTreeRegressor
dtree=DecisionTreeRegressor().fit(x_train,y_train)
```

```python
x_future = dataset[['Close']][:-future_days]
x_future=x_future.tail(future_days)
x_future=np.array(x_future)
x_future
```

```python
tree=dtree.predict(x_future)
tree
```

```python
shape=x.shape[0]
```

```python
dtree_confidence=dtree.score(x_future,tree)
dtree_confidence
```

```python
predictions=tree
valid=dataset[shape:]
valid['Predictions']=predictions
plt.figure(figsize=(12,4))
plt.title("Stock Price Prediction")
plt.xlabel("Days")
plt.ylabel("Price")
plt.plot(dataset['Close'])
plt.plot(valid[['Close' , 'Predictions']])
plt.legend(['Ori', 'val', 'Pred'])
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
lr_model=LinearRegression().fit(x_train,y_train)
```

```python
lr_pred=lr_model.predict(x_future)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
mse=mean_squared_error(lr_pred,x_future)
mse
lr_score=r2_score(x_future,lr_pred)
lr_score
```

```python
predictions=lr_pred
valid=dataset[shape:]
valid['Predictions']=predictions
plt.figure(figsize=(16,8))
plt.title("Stock Price Prediction")
plt.xlabel("Days")
plt.ylabel("Price")
plt.plot(dataset['Close'])
plt.plot(valid[['Close' , 'Predictions']])
plt.legend(['Original', 'values', 'Predicted'])
plt.show()
```
